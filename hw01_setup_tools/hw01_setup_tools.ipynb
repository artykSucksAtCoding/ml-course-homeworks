```python
import numpy as np
import pandas as pd
import sklearn
import sys

print(f"Python: {sys.version.split()[0]}")
print(f"NumPy: {np.__version__}")
print(f"Pandas: {pd.__version__}")
print(f"Scikit-learn: {sklearn.__version__}")
```

```python
X = np.random.randn(100, 5)
w = np.array([0.5, -0.2, 0.1, 0.8, -0.5])

# Статистики
print(f"Mean: {X.mean():.4f}, Std: {X.std():.4f}, Min: {X.min():.4f}, Max: {X.max():.4f}")
print(f"Mean by columns: {X.mean(axis=0)}")

# Векторизация
y = X @ w
```

```python
columns = ['feat_1', 'feat_2', 'feat_3', 'feat_4', 'feat_5']
df = pd.DataFrame(X, columns=columns)
df['target'] = y

print(df.head())
print(df.describe())
print(f"Missing values: {df.isnull().sum().sum()}")


print(df[df['target'] > 0].sort_values(by='target').head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 4))
sns.histplot(df['target'], kde=True)
plt.title("Distribution of Target")
plt.show()

plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```
